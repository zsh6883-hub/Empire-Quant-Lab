```python
import os
import sys
import time
import warnings
import pandas as pd
import numpy as np

# 全局高能降噪屏蔽
warnings.filterwarnings('ignore')

# ==========================================
# 🛠️ 核心防御：GitHub 大杀器极速免编译并网补丁
# ==========================================
try:
    import talib
    import mplfinance as mpf
    import akshare as ak
except ImportError:
    print("📡 [系统升级] 检测到核心大杀器未并网。启动 Python 3.12 专属二进制轮子矩阵...")

    # 使用社区预编译好的 TA-Lib 二进制包，绕过源码编译与 C 底层报错
    os.system('pip install ta-lib-bin > /dev/null 2>&1')
    # 同步安装视觉渲染总线与数据源
    os.system('pip install mplfinance akshare > /dev/null 2>&1')

    print("✨ [环境复活] 二进制组件并网成功。")
    print("💡 提示：Colab 环境需要刷新底层缓存。长官，【请点击 Colab 顶部菜单: 代码执行程序 -> 重启会话】，然后重新运行此单元格即可全线点火！")
    sys.exit(0)

# 帝国核心防御资产集群 (长官的5大主力战略标的)
PORTFOLIO_MAP = {
    "000725": "京东方A",
    "600157": "永泰能源",
    "601991": "大唐发电",
    "000767": "晋控电力",
    "000100": "TCL科技"
}

class EmpireUltimateTimeMatrixTerminal:
    """
    🏰 EMPIRE QUANT LAB - 二进制完备集成 + 历史时序矩阵滑动推演终端 (v3.7.8 Hotfix)
    """
    def __init__(self, initial_cash=1000000.0, max_positions=5, bias_limit=0.15):
        self.initial_cash = initial_cash
        self.available_cash = initial_cash
        self.max_positions = max_positions
        self.position_limit = initial_cash / max_positions
        self.bias_limit = bias_limit

        self.positions = {}
        self.trade_logs = []

    def fetch_and_build_global_matrix(self, lookback_days=180):
        """ 历史数据矩阵清洗与 TA-Lib 高阶指标注入 """
        print("📡 [链路并网] 正在拉取帝国核心标的的全局历史时序矩阵...")
        global_matrix = {}

        for code, name in PORTFOLIO_MAP.items():
            symbol_prefix = "sh" if code.startswith("60") else "sz"
            try:
                df_raw = ak.stock_zh_a_daily(symbol=f"{symbol_prefix}{code}", adjust="qfq")
                if df_raw is not None and not df_raw.empty:
                    df = df_raw[['date', 'open', 'high', 'low', 'close', 'volume']].copy()
                    df.columns = ['Date', 'Open', 'High', 'Low', 'Close', 'Volume']
                    df['Date'] = pd.to_datetime(df['Date'])
                    df.set_index('Date', inplace=True)

                    # 🧮 调用 TA-Lib 毫秒级原生底盘计算趋势防御线
                    close_prices = df['Close'].values
                    df['MA5'] = talib.SMA(close_prices, timeperiod=5)
                    df['MA20'] = talib.SMA(close_prices, timeperiod=20)
                    df['Bias20'] = (df['Close'] - df['MA20']) / df['MA20']

                    # 调用 TA-Lib 标准 MACD 动能矩阵
                    df['DIF'], df['DEA'], df['MACD'] = talib.MACD(close_prices, fastperiod=12, slowperiod=26, signalperiod=9)
                    df['MACD'] = df['MACD'] * 2
                    df['MACD_Slope'] = df['MACD'].diff()

                    # 💥 TA-Lib 智能形态识别：捕捉看涨/看跌吞没经典反转K线
                    df['K_Shape'] = talib.CDLENGULFING(df['Open'].values, df['High'].values, df['Low'].values, df['Close'].values)

                    global_matrix[code] = df.dropna()
            except Exception as e:
                print(f"⚠️ [接入受阻] {name}({code}) 数据同步发生异常: {e}")

        return global_matrix

    def render_wallstreet_chart(self, df, current_date, name, ticker, action_type="BUY"):
        """ 📊 华尔街级 mplfinance 高清走视图渲染 """
        slice_df = df.loc[:current_date].tail(45)
        if len(slice_df) < 20:
            return

        m_colors = mpf.make_marketcolors(up='#ff4d4d', down='#2ecc71', edge='inherit', wick='inherit', volume='inherit')
        empire_style = mpf.make_mpf_style(marketcolors=m_colors, gridcolor='#1e1e24', facecolor='#0e0e12', figcolor='#0a0a0c')

        ap_plots = [
            mpf.make_addplot(slice_df['MA5'], color='#ff9900', width=1.0),
            mpf.make_addplot(slice_df['MA20'], color='#00bcff', width=1.2),
            mpf.make_addplot(slice_df['MACD'], type='bar', color=['#ff4d4d' if m >= 0 else '#2ecc71' for m in slice_df['MACD']], panel=2, title="MACD Dynamic")
        ]

        date_str = current_date.strftime('%Y-%m-%d')
        print(f"🎬 [看板生成] 导出交割视觉复盘线 -> 标的: {name} | 动作: {action_type} | 时间点: {date_str}")

        mpf.plot(
            slice_df, type='candle', style=empire_style, addplot=ap_plots, volume=True,
            panel_ratios=(5, 2, 2), figscale=1.1,
            title=f"\nEMPIRE QUANT v3.7.8 - TRIGGER [{action_type}] AT {date_str}\nAsset: {name}({ticker}) | Close: {slice_df.iloc[-1]['Close']:.2f}"
        )

    def run_time_matrix_backtest(self, test_days=50):
        """ 🚀 启动时空隧道：自动化多周期历史轴滑动机制 """
        global_matrix = self.fetch_and_build_global_matrix()
        if not global_matrix:
            print("❌ 全局时序矩阵构建失败，中央控制总线熔断。")
            return

        common_dates = sorted(list(set.intersection(*[set(df.index) for df in global_matrix.values()])))
        timeline = common_dates[-test_days:]

        print(f"\n" + "🔥 " * 6 + f"中央回测矩阵点火成功 · 推演时序跨度: {timeline[0].strftime('%Y-%m-%d')} 至 {timeline[-1].strftime('%Y-%m-%d')}" + "🔥 " * 6)

        for current_date in timeline:
            date_str = current_date.strftime('%Y-%m-%d')

            # 1. 持仓资产风控巡检
            active_keys = list(self.positions.keys())
            for code in active_keys:
                name = PORTFOLIO_MAP[code]
                df = global_matrix[code]

                metrics = df.loc[current_date]
                price = float(metrics['Close'])
                ma20 = float(metrics['MA20'])
                macd = float(metrics['MACD'])
                macd_slope = float(metrics['MACD_Slope'])

                pos = self.positions[code]
                exit_trend = price < ma20
                exit_macd = macd < 0 and macd_slope < 0

                if exit_trend or exit_macd:
                    revenue = pos['shares'] * price
                    self.available_cash += revenue
                    pnl_pct = (price - pos['entry_price']) / pos['entry_price'] * 100
                    reason = "跌破生命线MA20" if exit_trend else "MACD动能死叉"

                    print(f"💥 [{date_str} 风控强平] {name}({code}) 触及防御红线({reason})！平仓价: {price:.2f} | 盈亏: {pnl_pct:+.2f}%")
                    self.render_wallstreet_chart(df, current_date, name=name, ticker=code, action_type="TREND_SELL")

                    self.trade_logs.append({'Date': date_str, 'Code': code, 'Name': name, 'Action': 'SELL', 'Price': price, 'Shares': pos['shares'], 'PnL': f"{pnl_pct:+.2f}%"})
                    del self.positions[code]

            # 2. 未持仓资产扫描
            for code, name in PORTFOLIO_MAP.items():
                if code in self.positions or len(self.positions) >= self.max_positions:
                    continue

                df = global_matrix[code]
                metrics = df.loc[current_date]

                price = float(metrics['Close'])
                ma20 = float(metrics['MA20'])
                bias20 = float(metrics['Bias20'])
                macd = float(metrics['MACD'])
                macd_slope = float(metrics['MACD_Slope'])
                k_shape = int(metrics['K_Shape'])

                condition_trend = price > ma20
                condition_momentum = macd > 0 or macd_slope > 0
                condition_shape = (k_shape == 100)

                if condition_trend and (condition_momentum or condition_shape):
                    if bias20 > self.bias_limit:
                        continue

                    shares_to_buy = int((self.position_limit / price) // 100) * 100
                    cost = shares_to_buy * price

                    if shares_to_buy > 0 and self.available_cash >= cost:
                        self.available_cash -= cost
                        self.positions[code] = {'name': name, 'shares': shares_to_buy, 'entry_price': price}
                        print(f"🟢 [{date_str} 开仓建仓] TA-Lib 矩阵多头共振成立！买入 {name}({code}) | 价格: {price:.2f} | 划转头寸: {cost:,.2f} RMB")
                        self.render_wallstreet_chart(df, current_date, name=name, ticker=code, action_type="MOMENTUM_BUY")

                        self.trade_logs.append({'Date': date_str, 'Code': code, 'Name': name, 'Action': 'BUY', 'Price': price, 'Shares': shares_to_buy, 'PnL': '0.00%'})

        self._print_final_statement(global_matrix, timeline[-1])

    def _print_final_statement(self, global_matrix, last_date):
        print("\n" + "👑 " * 10 + " 帝国大厦时序推演终极资产负债表 " + "👑 " * 10)
        allocated_funds = 0
        records = []
        for code, info in self.positions.items():
            last_price = float(global_matrix[code].loc[last_date]['Close'])
            val = info['shares'] * last_price
            allocated_funds += val
            records.append({
                '资产代码': code, '资产名称': info['name'], '持股数': f"{info['shares']:,}",
                '初始建仓价': info['entry_price'], '期末收盘价': last_price, '评估净值(RMB)': f"{val:,.2f}"
            })
        if records:
            print(pd.DataFrame(records).to_string(index=False))
        else:
            print("⚪ 审计结果：期末全线完美空仓避险，100% 现金护垫防御中。")
        print("-" * 80)
        print(f"🏦 期末账户可用现金储备: {self.available_cash:,.2f} RMB")
        print(f"👑 帝国集群实时最终总资产: {self.available_cash + allocated_funds:,.2f} RMB")
        print("👑 " * 30 + "\n")

if __name__ == "__main__":
    commander = EmpireUltimateTimeMatrixTerminal(initial_cash=1000000.0, max_positions=5, bias_limit=0.15)
    commander.run_time_matrix_backtest(test_days=50)
```
